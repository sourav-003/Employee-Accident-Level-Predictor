```python
from sklearn.feature_extraction.text import TfidfVectorizer
#Pip install nltk
from nltk import tokenize,stem
import nltk
nltk.download("vader_lexicon")
import nltk
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('stopwords')
from nltk.corpus import stopwords
# Sentiment intensity analyzer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
#pip install wordcloud
from wordcloud import WordCloud,STOPWORDS
```

```python
STOPWORDS.update(["kg","cm","am","pm","ore","mr","mrs"])
print(STOPWORDS)
```

```python
def nlp_preprocessor(row):
    sentence=row.Description
    #convert all characters to lowercase
    lowered=sentence.lower()
    tok=tokenize.word_tokenize(lowered)

    #Lematization & stemming
    lemmatizer=stem.WordNetLemmatizer()
    lem=[lemmatizer.lemmatize(i) for i in tok if i not in STOPWORDS]
    stemmer=stem.PorterStemmer()
    stems=[stemmer.stem(i) for i in lem if i not in STOPWORDS]

    # remove non-alphabetical characters like(":",".",",")
    alphas=[i for i in stems if i.isalpha() and (i not in STOPWORDS)]
    return " ".join(alphas)
```

```python
df["Description_processed"]=df.apply(nlp_preprocessor,axis=1)
df.head(2)
```

```python
def sentiment_calculate(text):
    analyzer=SentimentIntensityAnalyzer()
    sent_score=analyzer.polarity_scores(text)["compound"]

    return float(sent_score)
```

```python
df["Decription_sentiment_score"]=df["Description_processed"].apply(lambda x: sentiment_calculate(x) )
```

```python
import pandas as pd
import gradio as gr
import joblib
import warnings
import numpy as np
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk import tokenize, stem
from nltk.corpus import stopwords
from collections import defaultdict
import shap
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")

nltk.download('vader_lexicon', quiet=True)
nltk.download('punkt', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('stopwords', quiet=True)

lemmatizer = stem.WordNetLemmatizer()
analyzer = SentimentIntensityAnalyzer()


# --- 1. Load the Pre-trained Model and Encoders ---
try:
    # 📌 ACTION: Ensure these file paths are correct!
    rf_model = joblib.load('random_forest_model.joblib')
    encoders = joblib.load('label_encoders.joblib')
    data = pd.read_csv("case_study_employee_accident_in_plant.csv")
    print("Models, encoders, and data loaded successfully.")
except FileNotFoundError as e:
    raise FileNotFoundError(f"Missing essential file: {e}. Please ensure 'random_forest_model.joblib', 'label_encoders.joblib', and 'case_study_employee_accident_in_plant.csv' are in the same directory.")
except Exception as e:
    raise Exception(f"An error occurred during loading: {e}")

# --- 2. Define Preprocessing Variables and Functions ---
TRAINING_COLUMNS = [
    'Country', 'Local', 'Industry Sector', 'Gender', 'Employee type', 'critical',
    'Year', 'Month', 'Day', 'Weekday', 'WeekofYear', 'Season',
    'Decription_sentiment_score' # Must be the last column as it's numerical
]

STOPWORDS = set(stopwords.words('english'))
STOPWORDS.update(["kg", "cm", "am", "pm", "ore", "mr", "mrs"])

def nlp_preprocessor(sentence):
    """Cleans and processes text using lowering, tokenization, lemmatization, and stemming."""
    lowered = sentence.lower()
    tok = tokenize.word_tokenize(lowered)
    lem = [lemmatizer.lemmatize(i) for i in tok if i not in STOPWORDS]
    stemmer = stem.PorterStemmer()
    stems = [stemmer.stem(i) for i in lem if i not in STOPWORDS]
    alphas = [i for i in stems if i.isalpha() and (i not in STOPWORDS)]
    return " ".join(alphas)

def sentiment_calculate(text):
    """Calculates VADER compound sentiment score."""
    return float(analyzer.polarity_scores(text)["compound"])


# --- 3. Gradio Prediction Function ---

def predict_accident_level(Country, Local, Industry_Sector, Gender, Employee_type, critical, Year, Month, Day, Weekday, WeekofYear, Season, Description):

    # 1. Preprocess the description to get sentiment score
    processed_desc = nlp_preprocessor(Description)
    sentiment_score = sentiment_calculate(processed_desc)

    # 2. Create a DataFrame from the inputs
    input_data = pd.DataFrame({
        'Country': [Country], 'Local': [Local], 'Industry Sector': [Industry_Sector],
        'Gender': [Gender], 'Employee type': [Employee_type], 'critical': [critical],
        'Year': [int(Year)], 'Month': [int(Month)], 'Day': [int(Day)],
        'Weekday': [Weekday], 'WeekofYear': [int(WeekofYear)], 'Season': [Season],
        'Decription_sentiment_score': [sentiment_score]
    })

    # 3. Encode the categorical features using the LOADED encoders
    for col, le in encoders.items():
        if col in input_data.columns and col != 'Accident Level':

            # Function to safely encode an input value
            def safe_transform(value, encoder):
                """Transforms value using encoder, or returns the mode/a placeholder if unseen."""
                if value in encoder.classes_:
                    return encoder.transform([value])[0]
                else:
                    print(f"Warning: Unseen category '{value}' in column '{col}'. Encoded to -1.")
                    return -1

            input_data[col] = input_data[col].apply(lambda x: safe_transform(x, le))

    # 4. Ensure columns are in the same order as during training
    input_data = input_data[TRAINING_COLUMNS]

    # 5. Make the prediction (using predict_proba for better diagnosis)
    probabilities = rf_model.predict_proba(input_data)[0]

    # Get the index of the highest probability prediction
    predicted_index_encoded = np.argmax(probabilities)

    # Inverse transform the index to get the actual label
    predicted_level_label = encoders['Accident Level'].inverse_transform([predicted_index_encoded])[0]

    # 6. Prepare the final output strings
    all_encoded_levels = np.arange(len(probabilities)) # [0, 1, 2, 3, 4]
    all_levels_labels = encoders['Accident Level'].inverse_transform(all_encoded_levels)

    # Create the probability summary string
    prob_summary = defaultdict(float)
    for level, prob in zip(all_levels_labels, probabilities):
        prob_summary[level] = prob * 100

    summary_text = "\n".join([f"{level}: {prob_summary[level]:.2f}%" for level in sorted(prob_summary.keys())])

    return f"Predicted Accident Level: {predicted_level_label}", summary_text

# --- 4. Define the Gradio Interface ---

# Create lists of choices for the dropdowns from the original data
country_choices = list(data['Country_no'].unique())
local_choices = list(data['Local_status'].unique())
industry_choices = list(data['Industry Sector'].unique())
gender_choices = list(data['Sex'].unique())
employee_type_choices = list(data['Employee or Third Party'].unique())
critical_choices = list(data['Risk_assessment(critical)'].unique())

weekday_choices = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
season_choices = ['Spring', 'Summer', 'Autum', 'Winter']

# Define Gradio inputs
inputs = [
    gr.Dropdown(country_choices, label="Country"),
    gr.Dropdown(local_choices, label="Local"),
    gr.Dropdown(industry_choices, label="Industry Sector"),
    gr.Dropdown(gender_choices, label="Gender"),
    gr.Dropdown(employee_type_choices, label="Employee Type"),
    gr.Dropdown(critical_choices, label="Critical Risk"),
    gr.Number(label="Year (Numerical)", value=2024),
    gr.Slider(1, 12, step=1, label="Month (Numerical)", value=1),
    gr.Slider(1, 31, step=1, label="Day (Numerical)", value=1),
    gr.Dropdown(weekday_choices, label="Weekday"),
    gr.Slider(1, 53, step=1, label="Week of Year (Numerical)", value=1),
    gr.Dropdown(season_choices, label="Season"),
    gr.Textbox(label="Description of Event (for Sentiment Score)", lines=3, placeholder="Describe the incident..."),
]

# Define Gradio outputs
output = [
    gr.Textbox(label="Predicted Accident Level (Highest Probability)"),
    gr.Textbox(label="Full Class Probabilities (%)")
]

# Create and launch the interface
interface = gr.Interface(
    fn=predict_accident_level,
    inputs=inputs,
    outputs=output,
    title="Employee Accident Level Predictor",
    description="The model predicts the most likely accident level. Check the Full Class Probabilities to see the severity of the class imbalance (most probability will likely be on Level I or II)."
)

if __name__ == '__main__':
    interface.launch()
```
